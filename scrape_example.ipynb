```python
import requests
with open("example.html", 'w') as example:
    example.write(requests.get("https://raw.githubusercontent.com/gerzytet/web-scraping-workshop-files/refs/heads/main/example.html").text)
```

```python
from bs4 import BeautifulSoup
example_doc = open('example.html', encoding='utf-8').read()
print(example_doc)
soup = BeautifulSoup(example_doc, 'html.parser')
```
